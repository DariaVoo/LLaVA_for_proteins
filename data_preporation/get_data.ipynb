```python
from Bio import SeqIO
```

```python
seq_records = {}  # keys remain ordered since Python 3.7 :)
seq_records_no_function = []

i = 0
with open(src_path) as input_handle:
    for record in SeqIO.parse(input_handle, "uniprot-xml"):
        seq = str(record.seq)
        if "comment_function" in record.annotations.keys():
            description = record.annotations["comment_function"]
            seq_records[record.id] = (seq, description)
        else:
            print(record)
            seq_records_no_function.append(seq)

        i+=1
        if i == 50:
            break
```
